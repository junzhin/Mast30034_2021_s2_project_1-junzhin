# 2. Data Cleaning

```python
# passenger_count is supposed to be bewteen 1 and 7
yellow_01_df = yellow_01_df.loc[(yellow_01_df["passenger_count"] != 0)& (yellow_01_df["passenger_count"] <= 7),:].sort_values("passenger_count")
yellow_01_df
```

```python
yellow_01_df = yellow_01_df.loc[(yellow_01_df["passenger_count"] != 0),:].sort_values("passenger_count")
yellow_01_df = yellow_01_df.loc[(yellow_01_df["passenger_count"] <= 7),:].sort_values("passenger_count")
yellow_01_df
```

```python
# invalid payment type 3,4,5
yellow_01_df = yellow_01_df.loc[(yellow_01_df["payment_type"] != 3) ,:]
yellow_01_df
```

```python
#  the minium fare amout is above 2.5 
yellow_01_df = yellow_01_df.loc[(yellow_01_df["fare_amount"] >=2.5) ,:]
yellow_01_df
```

```python
#  the minium total amout is above 2.5 
yellow_01_df = yellow_01_df.loc[(yellow_01_df["total_amount"] >=2.5) ,:]
yellow_01_df
```

```python
#  the total distance is above 0
yellow_01_df = yellow_01_df.loc[(yellow_01_df["trip_distance"] > 0) ,:]
yellow_01_df
```

# 3 Remove not_interested features

```python
yellow_01_df = yellow_01_df.drop(['VendorID','store_and_fwd_flag','congestion_surcharge','improvement_surcharge','tolls_amount','mta_tax','extra'], axis=1)
yellow_01_df
```
